# 面向对象引子——人狗大战

人狗大战

- 多条狗，每个狗有名字，品种，攻击力
- 可以有多个人 ，
- 狗可以咬人，人可以打狗

```python
# 多条狗，重复性太多

dog = {
    "name":"zsy",
    "d_type":"藏獒",
    "attack_val":80,
    "life_val":100,
}

dog = {
    "name":"zgy",
    "d_type":"哈士奇",
    "attack_val":30,
    "life_val":100,
    
}
```

```python
### 一条条狗定义太重复了，写进一个函数里


# ----------------------------模板-------------------------
# 攻击力与类型对应好
attack_val = {
    '哈士奇':30,
    '藏獒':80
}

# 定义狗
def dog(name,d_type):
  data = {
      "name":name,
      "d_type":d_type,
      "life_val":100
  }

  #根据狗的类型赋予攻击力
  if d_type in attack_val:
    data['attack_val'] = attack_val[d_type] #字典通过这种方式可以直接加一项
  else:
     data['attack_val'] = 15 #比如说中华田园犬

  return data
  
# 定义人
def person(name,age):
  data = {
      "name":name,
      "age":age,
      "life_val":100
  }

  if age>18:
    data['attack_val'] = 50
  else:
    data['attack_val'] = 30
  
  return data

# 哪知狗咬了哪个人一口
def dog_bite(dog_obj, person_obj):
  person_obj['life_val'] -= dog_obj['attack_val'] #执行咬人动作
  print('狗{}咬了人{}一口，人掉血{}，还有血量{}...'.format(dog_obj['name'],
                              person_obj['name'],
                              dog_obj['attack_val'],
                              person_obj['life_val']))

# 哪个人打了哪个人一下
def beat(person_obj, dog_obj):
  dog_obj['life_val'] -= person_obj['attack_val']
  print('人{}打了狗{}一下，狗掉血{}，还有血量{}...'.format(person_obj['name'],
                              dog_obj['name'],
                              person_obj['attack_val'],
                              dog_obj['life_val']))


#----------------------具体的实体-----------------------
d1 = dog('大黄','藏獒')
d2 = dog('小黑','哈士奇')

p1 = person('rc',16)
p2 = person('gy',20)


print(d1,d2)
print(d1["attack_val"])
print(d2["attack_val"])

print(p1,p2)
print(p1["attack_val"])
print(p2["attack_val"])

dog_bite(d1,p1)
beat(p2,d2)

# 这样调用是错误的，但是程序没有报错，看下面怎么改
dog_bite(p1,d1)
```

```python
### 让狗才能调用狗咬人的函数，把咬人函数写进定义狗的函数里面


# ----------------------------模板-------------------------
# 攻击力与类型对应好
attack_val = {
    '哈士奇':30,
    '藏獒':80
}

# 定义狗
def dog(name,d_type):
  data = {
      "name":name,
      "d_type":d_type,
      "life_val":100
      # "bite"
  }

  #根据狗的类型赋予攻击力
  if d_type in attack_val:
    data['attack_val'] = attack_val[d_type] #字典通过这种方式可以直接加一项
  else:
     data['attack_val'] = 15

  # 哪知狗咬了哪个人一口
  def dog_bite(person_obj):
    person_obj['life_val'] -= data['attack_val'] #执行咬人动作
    print('狗{}咬了人{}一口，人掉血{}，还有血量{}...'.format(data['name'],
                                person_obj['name'],
                                data['attack_val'],
                                person_obj['life_val']))
  
  data['bite'] = dog_bite # 这里不加括号，说明定义狗时没有调用，但是能够使得从外部访问

  return data
  
# 定义人
def person(name,age):
  data = {
      "name":name,
      "age":age,
      "life_val":100
      # "beat"
  }

  if age>18:
    data['attack_val'] = 50
  else:
    data['attack_val'] = 30
  
  # 哪个人打了哪个人一下
  def beat(dog_obj):
    dog_obj['life_val'] -= data['attack_val']
    print('人{}打了狗{}一下，狗掉血{}，还有血量{}...'.format(data['name'],
                                dog_obj['name'],
                                data['attack_val'],
                                dog_obj['life_val']))
  
  data['beat'] = beat

  return data





#----------------------具体的实体-----------------------
d1 = dog('大黄','藏獒')
d2 = dog('小黑','哈士奇')

p1 = person('rc',16)
p2 = person('gy',30)


print(d1,d2)
print(p1,p2)

d2['bite'](p1) #先找到然后调用
d2['bite'](p1) #先找到然后调用
d2['bite'](p1) #先找到然后调用

p2['beat'](d1)
p2['beat'](d1)

d1['bite'](p2)
d1['bite'](p2)
```

说了这么多，这跟面向对象有什么关系么？ 当然有，其实你上面写的代码，就是面向对象的代码。

你在设计角色时，为了让一个角色可以变成多个实体对象，你设计了一个基础模板，只要传入不同参数，就会产生不同的狗。 这代表你已经开始切换成上帝视角看事情 ，上帝视角就是面向对象编程的视角，上帝要造世界万物，他肯定不是一个一个的造出来，他肯定是设计出一个个的物种的模板，然后通过模子批量批一个个的实体造出来。造出来的实体各有特色，属性、功能都不尽相同，有的人的贪婪、有的人好色、有的人懦弱，有的人勇猛。这些人之间会发生什么关系 ，谁和谁交媾、谁和谁打仗，上帝懒的管，上帝只控制大局。听着玄乎吧，我们接下来一点点接晓怎么通过面向对象在编程世界里做上帝。
