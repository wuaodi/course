# isinstance\issubclass

看到了知道就行

````python
# isinstance(obj,cls) 检查obj是否是类cls的对象

class Foo(object):
  pass

obj = Foo()

isinstance(obj,Foo)
````

````python
# issubclass(cls1,cls2) 检查cls1是否是类cls2的子类

class Foo(object):
  pass

class Bar(Foo):
  pass

issubclass(Bar,Foo)
````

except动词的意思是除了……，exception名词的意思就是异常

Exception 是万能异常，所有错误的鼻祖，能够捕捉各种异常

```
try:
  你的代码

  “你的代码”出错后要执行的代码
```



````python
while True:
  num1 = int(input("n1>:").strip())
  num2 = int(input("n2>:").strip())

  res = num1 + num2
  print('result:',res)

# 如果输入的不是数字，会打印出一堆的报错信息
````

````python
while True:
  try:
    num1 = int(input("n1>:").strip())
    num2 = int(input("n2>:").strip())

    res = num1 + num2
    print('result:',res)
  except Exception as e: #把异常提示传给临时变量e
    print(e)
    print('亲，要不要检查一下你输入的是不是数字？')

# 这个时候就不会打印一堆的报错信息了，而是会只打印出最后的错误提示语句，和你相对用户说的话
````

````python
### 对异常的更加细化的处理

while True:
  try:
    num1 = int(input("n1>:").strip())
    num2 = int(input("n2>:").strip())
    res = num1 + num2

    print('result:',res)
    # print('result:',res,csu) #这里是NameError，csu这个变量并没有被定义

    open("一个秘密.txt")

  except ValueError as e: #不是用万能异常，而是更加精细化地捕捉异常
    print(e)
    print('亲，要不要检查一下你输入的是不是数字？')

  except NameError as e:
    print(e)

  except FileNotFoundError as e:
    print(e)5
````

# 自定义异常

## else finally 万能异常

````python
while True:
  try:
    num1 = int(input("n1>:").strip())
    num2 = int(input("n2>:").strip())
    res = num1 + num2

    print('result:',res)
    # print('result:',res,csu) #这里是NameError，csu这个变量并没有被定义

    # open("一个秘密.txt") #这里是FileNotFoundError


  except ValueError as e:
    print(e)
    print('亲，要不要检查一下你输入的是不是数字？')

  except NameError as e:
    print(e)

  except FileNotFoundError as e:
    print(e)

  except Exception as e: #万能异常放在最后，用来捕捉没有预料到的异常
    print("发生错误",e)

  else:
    print("try里面的代码没有异常走这里")

  finally:
    print("不管有没有发生异常，都走这里")
````

````python
class YoutubeConnectionError(BaseException): #继承最基本的异常类
  def __init__(self,msg):
    self.msg = msg
  def __str__(self):
    return self.msg

try:
  print("正在连接www.youtube.com......")
  raise YoutubeConnectionError("出现错误：在中国无法访问Youtube")
except YoutubeConnectionError as e:
  print(e)
````

## assert 断言

用于判断代码的执行是否符合预期



```
assert 判断语句, "提示信息"
```

如果判断语句结果为True，继续执行；

如果判断语句结果为False，中断执行，返回“提示信息”


````python
### 用于判断运行结果是否符合预期
a = 1
b = 2
assert a != b #结果为True不报错
assert a == b #结果为False报错
````

````python
### 用于判断输入数据格式是否正确
def hahaha(name,age):
  assert type(name) is str, "输入数据类型不正确"
  assert type(age) is int, "输入数据类型不正确"

hahaha('zgy',18)
hahaha('zgy','18')
````
