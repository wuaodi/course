# 省流版本：

**普通方法**在s.func()调用的时候，自动传进去的是**实例本身**作为self

**类方法**在s.func()调用的时候，自动传进去的是**类本身**作为self

**静态方法**在s.func()调用的时候，**不自动传进去**任何内容

**属性方法**调用的时候不加括号，s.func

# 类方法 @clsaamethod

类方法通过@classmethod装饰器实现，类方法和普通方法的区别是， 类方法只能访问类变量，不能访问实例变量

因为加了@classmethod传进来的self是类本身，而不是实例本身

```python
class Dog(object):
  def __init__(self,name):
     self.name = name

  @classmethod # 类方法不能访问实例变量
  def eat(self):
    print(self.name,'在吃东西')

d = Dog('大黄')
d.eat()
```

```python
class Dog(object):
  name = '一条蠢狗' # 类方法能够访问类变量

  def __init__(self,name):
     self.name = name

  @classmethod
  def eat(self):
    print('打印一下发现类方法传进来的是类本身：',self) # 类方法传进来的self是类本身
    print(self.name,'在吃东西')

d = Dog('大黄')
d.eat()
print(Dog) # 与类方法中的self一致
```

```python
class Dog(object):
  name = '一条蠢狗'

  def __init__(self,name):
     self.name = name

  def eat(self):
    print('不加classmethod传进来的self是类对象：',self)
    print(self.name,'在吃东西')

d = Dog('大黄')
d.eat()
```

有什么用？

其实没有太多用，自己写的时候几乎用不到，看别人的代码有时候会有

# 静态方法 @staticmethod

通过@staticmethod装饰器即可把其装饰的方法变为一个静态方法，静态方法是不可以访问实例变量、类变量或类内其他方法，在通过实例调用的时候，s.func()，不自动传进来实例本身或类本身

静态方法隔断了它跟类或者实例的任何关系，只是放在类下的一个函数

```python
class Student(object):
  role = 'student'

  def __init__(self,name):
    self.name = name

  @staticmethod
  def fly(self):
    print(self.name,'正在天上飞……')

  @staticmethod #与类完全隔断，没有啥关系
  def add(a,b):
    print(a+b)


s1 = Student('张三')
# 不会自动执行s1.fly(s1)了，就是个能通过类名或实例名调用的普通函数
# s1.fly() #TypeError: fly() missing 1 required positional argument: 'self'

# 需要通过手动给函数参数
s1.fly(s1)
s1.add(5,8)
```

# 属性方法 @property

定义：属性方法的作用就是通过@property把一个方法变成一个静态属性（变量）

除了调用的时候有区别，别的任何都与普通方法一样

```python
class Student(object):
  role = 'student'

  def __init__(self,name):
    self.name = name

  def check_height(self):
    # 查高度的操作,假设返回值为500
    return 500

  @property
  def fly_height(self):
    print(self.name,'正在天上飞……')
    # 调用一个查高度的函数，高度是一个不断变化的值
    height = self.check_height()
    print('目前飞行高度为:',height)

s = Student('zgy')

s.fly_height # 调用的时候不用加括号，像是调用一个变量一样
```
