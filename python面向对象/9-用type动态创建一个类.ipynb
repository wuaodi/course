```python
# 一切皆对象，实例是类的对象，类也是对象，由某个东西生出来的

class Person(object):
  def __init__(self,name,age):
    self.name = name
    self.age = age

p = Person('zgy',18)

print(type(p)) #源于__mian__文件下的Person类
print(type(Person)) #怎么出现了一个type？type其实是所有类的祖师爷，类是由type产生的
print(type(type)) #type是由type产生的，套娃
```

```python
# 使用type产生类
Dog = type('Dog',(object,),{'role':'一只狗'})

# 实例化
d = Dog()
print(d.role)
```

```python
# 使用type产生一个复杂点的类
def __init__(self,name,age):
  self.name = name
  self.age = age

Dog = type('Dog',(object,),{'role':'一只狗','__init__':__init__})


# 实例化
d = Dog('大黄',3)
print(d.name)
```
