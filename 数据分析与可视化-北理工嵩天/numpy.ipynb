# numpy的数组对象：ndarray


numpy提供了：

1. 一个强大的表示N维数组的对象 ndarray
2. 广播函数功能，**用于数组间计算**
3. 整合C/C++/Fortran代码的工具
4. **线性代数、傅里叶变换、随机数生成等功能**

NumPy 是 SciPy、Pandas 等数据处理库的基础






```python
# numpy的引用
import numpy as np


# 基本语法中有了列表类型，为什么需要额外的数据类型？
# 举例：算A**2 + B**3，其中A和B是一维数据


# 基本python语法实现
def pySum():
  a = [0, 1, 2, 3, 4]
  b = [9, 8, 7, 6, 5]
  c = []
  for i in range(len(a)):
    c.append(a[i]**2 + b[i]**3)
  return c

print(pySum())


# 使用numpy实现
import numpy as np
def npSum():
  a = np.array([0, 1, 2, 3, 4])
  b = np.array([9, 8, 7, 6, 5])
  c = a**2 + b**3
  return c

print(npSum())


# 可见，使用numpy有两个很突出的优点：
# 1、数组对象可以去掉元素间运算所需的循环，使一维向量更像单个数据
# 2、numpy底层会加速数组对象的运算速度，底层是用C完成的
# 3、科学计算中，一个维度的数据类型往往是相同的
```

```python
# N维数组对象 ndarray

# 由两部分构成：
# 1、实际的数据
# 2、这些数据的描述（数据维度、数据类型等）

# ndarray要求元素类型相同，数组下标从0开始
# ndarray在程序中就是np.array()创建

# 轴（axis）：保存数据的维度
# 秩（rank）：轴的数量

# ndarray对象有5个属性：
# .ndim 秩，即轴的数量或维度的数量
# .shape 对象的尺度，对于矩阵，n行m列
# .size 对象的元素个数，相当于.shape中n*m的值
# .dtype 对象的元素类型
# .itemsize 对象中每个元素的大小，以字节为单位

import numpy as np

a = np.array([[0, 1, 2, 3, 4],
        [9, 8, 7, 6, 5]])

print(a.ndim)
print(a.shape)
print(a.size)
print(a.dtype)
print(a.itemsize)


# 非同质的ndarray，即每个维度包含的元素数量或元素类型不同
# 尽量避免使用，无法发挥numpy的优势
a = np.array([[0, 1, 2, 3, 4],
        [9, 8, 7, 6]])

print('\n')
print(a.ndim)
print(a.shape)
print(a.size)
print(a.dtype)
print(a.itemsize)
```

# ndarray数组的创建
- 从python的列表、元组等类型创建ndarray数组
- 使用numpy提供的函数创建ndarray数组，最常用
- 使用NumPy中其他函数创建ndarray数组

```python
### 从python的列表、元组等类型创建ndarray数组

# x = np.array(list/tuple)
# x = np.array(list/tuple, dtype=np.float32) 创建时指定元素的类型，不指定则程序自动确定

import numpy as np

# 从列表类型创建
x = np.array([0, 1, 2 ,3])
print(x)

# 从元组类型创建
x = np.array((4, 5, 6, 7))
print(x)

# 从列表和元组混合类型创建，列表和元组包含的元素数目相同即可
x = np.array([[1, 2], [9, 8], (0.1, 0.2)])
print(x)
```

```python
### 使用numpy提供的函数创建ndarray数组，最常用

# np.arange(n) 类似range()函数，返回ndarray类型，元素从0到n‐1
# np.ones(shape) 根据shape生成一个全1数组，shape是元组类型
# np.zeros(shape) 根据shape生成一个全0数组，shape是元组类型
# np.full(shape,val) 根据shape生成一个数组，每个元素值都是给定的val值
# np.eye(n) 创建一个正方的n*n单位矩阵，对角线为1，其余为0
# 大多数默认创建的是浮点数类型，也可以在创建时使用dtype指定

# np.ones_like(a) 根据数组a的形状生成一个全1数组
# np.zeros_like(a) 根据数组a的形状生成一个全0数组
# np.full_like(a,val) 根据数组a的形状生成一个数组，每个元素值都是val



import numpy as np

a = np.arange(10)
b = np.ones((3,6))
c = np.zeros((3,6), dtype=np.int32)
d = np.full((3,6), 100)
e = np.eye(5)

print('\n'), print(a)
print('\n'), print(b)
print('\n'), print(c)
print('\n'), print(d)
print('\n'), print(e)
print('\n')
print('--------------------------------------')

f = np.ones_like(a)
print('\n'), print(f)
g = np.zeros_like(a)
print('\n'), print(g)
h = np.full_like(a,581)
print('\n'), print(h)
print('\n')
print('-------------------------------------')

# 创建多维数组
x = np.ones((2, 3, 4))
print('\n')
print(x)
print(x.shape)
```

```python
### 使用NumPy中其他函数创建ndarray数组
# np.linspace() 根据起止数据等间距地填充数据，形成数组
# np.concatenate() 将两个或多个数组合并成一个新的数组
import numpy as np

a = np.linspace(1, 10, 4) # 从1到10，取4个数
b = np.linspace(1, 10, 4, endpoint=False) # 从1到10，取4个数，不包含最后一个数10
c = np.concatenate((a,b))

print('\n'), print(a)
print('\n'), print(b)
print('\n'), print(c)
print('\n')
```

# ndarray数组的变换
- 数组维度变换
- 元素类型变换

```python
### 维度变换
# .reshape(shape) 不改变数组元素，返回一个shape形状的数组，原数组不变
# .resize(shape) 与.reshape()功能一致，但修改原数组，无返回值(减少使用)
# .swapaxes(ax1,ax2) 将数组n个维度中两个维度进行调换，原数组不变
# .flatten() 对数组进行降维，返回折叠后的一维数组，原数组不变

import numpy as np



print('-------------reshape-------------')
a = np.ones((2,3,4), dtype=np.int)
print('a = \n',a), print('\n')
b = a.reshape((3,8))
print('a = \n',a), print('\n')
print('b = \n',b), print('\n')

print('-------------resize-------------')
a = np.ones((2,3,4), dtype=np.int)
print('a = \n',a), print('\n')
c = a.resize((3,8))
print('a = \n',a), print('\n')
print('c = \n',c)

print('-------------swapaxes-------------')
a = np.ones((2,3,4), dtype=np.int)
print('a = \n',a), print('\n')
b = a.swapaxes(0,1)
print('a = \n',a), print('\n')
print('b = \n',b), print('\n')


print('-------------flatten-------------')
a = np.ones((2,3,4), dtype=np.int)
print('a = \n',a), print('\n')
b = a.flatten()
print('a = \n',a), print('\n')
print('b = \n',b), print('\n')

print('Run successfully!')
```

```python
### 类型变换
# new_a = a.astype(new_type) 数组内元素的类型变换，返回一个新的值，原数组不变
# ls = a.tolist() 数组向列表的转换，原数组不变

import numpy as np

a = np.ones((2,3,4), dtype=np.int)
print(a),print('\n')

b = a.astype(np.float)
print(b),print('\n')
print(a),print('\n')

c = a.tolist()
print(c),print('\n')
print(a),print('\n')

print('Run successfully!')
```

# 数据的CSV文件存取（一维、二维

CSV Comma-Seperated Value

存取**一维或二维数组**方便的格式

np.savetxt()和np.loadtxt()只能有效存取一维和二维数组


**将numpy数组存入csv文件可以用 np.savetxt**


np.savetxt(frame, array, fmt='%.18e', delimiter=None)

• frame : 文件、字符串或产生器，可以是.gz或.bz2的压缩文件，例如xxx.csv

• array : 存入文件的数组

• fmt : 写入文件的格式，例如：%d %.2f %.18e，默认是18位小数的浮点数

• delimiter : 分割字符串，默认是任何空格，csv要改为逗号

```python
import numpy as np
a = np.arange(100).reshape(5,20)
```

```python
np.savetxt('a.csv',a,fmt='%d',delimiter=',')
```

```python
np.savetxt('a.csv',a,fmt='%.2f',delimiter=',')
```

**读入csv文件到numpy数组**

np.loadtxt(frame, dtype=np.float, delimiter=None， unpack=False)


• frame : 文件、字符串或产生器，可以是.gz或.bz2的压缩文件，即文件名，xxx.csv

• dtype : 数据类型，可选(默认是float，一般不用管，都进来之后再说)

• delimiter : 分割字符串，默认是任何空格，对于csv文件要用逗号

• unpack : 如果True，读入属性将分别写入不同变量


```python
b = np.loadtxt('a.csv',delimiter=',')
```

```python
b
```

```python
# 如果a.csv里面存储的是带小数的，指定int来读会出错
b = np.loadtxt('a.csv',dtype=np.int,delimiter=',')
```

# 多维数据存取

**存**

• frame : 文件、字符串

• sep : 数据分割字符串，如果是空串，写入文件为二进制

• format : 写入数据的格式


```python
import numpy as np
a = np.arange(100).reshape(5,2,10)
```

```python
a
```

```python
# 保存为一个由comma分割的文件
a.tofile("b.dat",sep=',',format='%d')
```

```python
# 默认保存为一个二进制的文件，占用空间更小
a.tofile("c.dat",format="%d")
```

**取**

np.fromfile(frame, dtype=float, count=‐1, sep='')

• frame : 文件、字符串

• dtype : 读取的数据类型

• count : 读入元素个数，‐1表示读入整个文件

• sep : 数据分割字符串，如果是空串，写入文件为二进制

```python
b = np.fromfile("b.dat",dtype=np.int,sep=',')
```

```python
# 失去了维度信息
b
```

```python
# 读取的时候恢复维度信息
b = np.fromfile("b.dat",dtype=np.int,sep=',').reshape(5,2,10)
```

```python
b
```

# “图像的手绘效果”实例分析

手绘效果的几个特征：

• 黑白灰色

• 边界线条较重

• 相同或相近色彩趋于白色

• 略有光源效果

实现以上特征需要以下技术：

1、 求梯度，梯度重构

2、 光源效果的实现

3、 梯度与光源相互作用，将梯度转化为灰度

4、 图像生成

```python
# 下载测试图片'beijing.jpg'
!pip install gdown
!gdown https://drive.google.com/uc?id=1blKdTQxj2O69uRfeYuKv_P_uFmT6bIn3
```

```python
from PIL import Image
import numpy as np

a = np.array(Image.open('beijing.jpg').convert('L')).astype('float')

# 梯度是灰度图像的变化率，可以用于增强边缘
# 立体效果通过虚拟深度值来实现，虚拟深度可以理解为凸出纸面的最高深度，想像成从上往下拍


# 利用虚拟深度和梯度值对图像进行重构

# 求梯度
depth = 10.0 #取值范围(0-100)
grad = np.gradient(a) #灰度图像的梯度值
grad_x, grad_y = grad
grad_x = grad_x * (depth/100.0) #添加深度对梯度的影响因素
grad_y = grad_y * (depth/100.0) #depth/100是将其归一化到0-1之间
grad_z = 1.0 #这个是一个假设值

# 将梯度归一化到0-1之间
A = np.sqrt(grad_x**2 + grad_y**2 + 1)
uni_x = grad_x/A
uni_y = grad_y/A
uni_z = 1/A

# 模拟光源
vec_el = np.pi/2.2
vec_az = np.pi/4.0
dx = np.cos(vec_el)*np.cos(vec_az)
dy = np.cos(vec_el)*np.sin(vec_az)
dz = np.sin(vec_el)

# 梯度与光源相互作用
b = 255*(dx*uni_x + dy*uni_y + dz*uni_z)
b.clip(0,255)
im = Image.fromarray(b.astype('uint8'))
im.save('beijingHD.jpg')
```
